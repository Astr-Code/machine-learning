```python
#导入sklearn.ensemble.RandomForestRegressor, train_test_split, matplotlib, numpy, pandas模块
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
plt.rcParams['font.sans-serif'] = ["SimHei"]
```

```python
#处理数据
data = pd.read_csv(r'F:\vscodefiles\ML\000001.csv', encoding='gbk')
X, y = data.loc[:], data.loc[:, "涨跌幅"]
X = X.drop(["涨跌幅"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print(X_test.isnull().any().any())
print(X_train.isnull().any().any())
```

```python
trees = RandomForestRegressor(n_estimators = 1000).fit(X_train, y_train)
# 模型评估
print("Training set score: {:.3f}".format(trees.score(X_train, y_train)))
print("Test set score: {:.3f}".format(trees.score(X_test, y_test)))
```

```python
print(trees.score(X, y))
```

```python
#特征重要性可视化
def plot_feature_importances_cancer(model):
    n_features = X_train.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), X_train.columns)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
plot_feature_importances_cancer(trees)
```
